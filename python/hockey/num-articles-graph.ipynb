```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the data
file_path = './elias_google_search_results.csv'
data = pd.read_csv(file_path)

# Convert 'Start Date' to datetime format
data['Start Date'] = pd.to_datetime(data['Start Date'])

# Plot the data
plt.figure(figsize=(12, 8))
plt.plot(data['Start Date'], data['Num Results'], marker='o', linestyle='-', color='b')

# Enhance x-axis labels and grid
plt.title('Number of Articles Posted about Elias Pettersson Per Week Over Time')
plt.xlabel('Week Start Date')
plt.ylabel('Number of Articles')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(rotation=45)

# Add more x-axis labels
plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator())
plt.gca().xaxis.set_minor_locator(mdates.DayLocator(interval=1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))

# Set x-axis limits to start and end at the first and last data points
plt.xlim(data['Start Date'].min(), data['Start Date'].max())

plt.tight_layout()

# Show the plot
plt.show()
```
